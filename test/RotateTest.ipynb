```python
%%ignite

# Rotate 2 pi radians. Expect it to look the same as rotating 0 degrees

def setup():
    size(300, 300)
    background("yellow")
    
    # Draw a non-rotated rectangle - should appear as border to rotated rect
    fill_style("white")
    rect(95, -5, 90, 60)
    
    rotate(2 * pi)
    
    # Expect black rectangle to land perfectly on top of white rectangle
    fill_style("black")
    rect(100, 0, 80, 50)
```

```python
%%ignite

# Rotate negative radians. Expect rotating -x radians == 2pi - x radians

def setup():
    size(300, 300)
    background("yellow")
    
    angle = pi / 6  # 30 degrees
    
    # Draw a positively rotated rect - should appear as border to rotated rect
    rotate(2 * pi - angle)
    fill_style("white")
    rect(95, -5, 90, 60)
    rotate(-(2 * pi - angle))
    
           
    rotate(-angle)
    
    # Expect black rectangle to land perfectly on top of white rectangle
    fill_style("black")
    rect(100, 0, 80, 50)
```
