```python
%%ignite
# Unit test: keys_held

# Note that this test is keyboard-dependent: if your keyboard layout differs from the US QWERTY layout, this will fail. 
# Also, keyboards without n-key rollover will likely experience jamming (i.e. some keys don't show as pressed)
def setup():
    size(500, 500)

def draw():
    background("default")
    draw_key("w", 225, 50, 40, 5)
    draw_key("a", 175, 100, 40, 5)
    draw_key("s", 225, 100, 40, 5)
    draw_key("d", 275, 100, 40, 5)
    
def draw_key(key, x, y, k_size, offset):
    fill_style(255, 255, 255, 0.5)
    if keys_held(key):
        fill_style("white")
    square(x + offset, y + offset, k_size)
    fill_style("black")
    text_size(int(k_size*0.8))
    text(key.upper(), x + offset + k_size*0.1, y + offset + k_size*0.1)
        
```

```python
%%ignite

def setup():
    size(400,400)

ball_x = {"x": 200, "y": 200, "z": 0}
ball_v = {"x": 0, "y": 0, "z": 0}
ball_r = 30
n_boxes = 0
def draw():
    global ball_x, ball_v, ball_r, n_boxes
    
    background("default")
    
    fill_style("white")
    stroke_style("black")
    for i in range(n_boxes):
        square(10 * (i%(width//10)), 10 * (i//(width//10)), 10)
    
    ball_apparent_r = 100*ball_r / (100 + ball_x["z"])
    fill_style(0, 0, 0, 0.2)
    fill_circle(ball_x["x"], ball_x["y"], 2*ball_r)
    fill_style("red")
    circle(ball_x["x"], ball_x["y"], 2*ball_apparent_r)
    
    if keys_held("ArrowUp"):
        fill_style(0,0,0,0.8)
    else:
        fill_style(0,0,0,0.4)
    fill_triangle(50, 40, 70, 20, 90, 40)
    
    if keys_held("ArrowDown"):
        fill_style(0,0,0,0.8)
    else:
        fill_style(0,0,0,0.4)
    fill_triangle(50, 100, 70, 120, 90, 100)
    
    if keys_held("ArrowRight"):
        fill_style(0,0,0,0.8)
    else:
        fill_style(0,0,0,0.4)
    fill_triangle(100, 50, 120, 70, 100, 90)
    
    if keys_held("ArrowLeft"):
        fill_style(0,0,0,0.8)
    else:
        fill_style(0,0,0,0.4)
    fill_triangle(40, 50, 20, 70, 40, 90)
    
    if keys_held("ArrowUp", "ArrowDown", pattern=[True, False]):
        ball_v["y"] = -2
    elif keys_held("ArrowUp", "ArrowDown", pattern=[False, True]):
        ball_v["y"] = 2
    else:
        ball_v["y"] = 0
    
    if keys_held("ArrowRight", "ArrowLeft", pattern=[True, False]):
        ball_v["x"] = 2
    elif keys_held("ArrowRight", "ArrowLeft", pattern=[False, True]):
        ball_v["x"] = -2
    else:
        ball_v["x"] = 0
    
    if ball_x["z"] + ball_v["z"] <= 0:
        ball_v["z"] = -ball_x["z"]
    else:
        ball_v["z"] -= 0.2
    
    ball_x["x"] += ball_v["x"]
    ball_x["y"] += ball_v["y"]
    ball_x["z"] += ball_v["z"]

def key_pressed():
    global n_boxes, ball_x, ball_v
    if key == "+":
        n_boxes += 1
    elif key == " " and ball_x["z"] == 0:
        ball_v["z"] = 10

def key_repeated():
    global n_boxes
    if key == "+":
        n_boxes += 1

def key_released():
    global n_boxes, ball_x, ball_v
    if key == "Escape":
        n_boxes = 0
        ball_x["x"], ball_x["y"], ball_x["z"] = 200, 200, 0
        ball_v["x"] = ball_v["y"] = ball_v["z"] = 0
        
```

```python
%%ignite
def setup():
    size(500, 150)

typed_str = ""
cursor_pos = 0
str_pos = 0
max_chars = 21
    
def draw():
    background("white")
    stroke_rect(50, 50, 400, 50)
    canvas.font="32px monospace"
    shown_str = typed_str[str_pos:str_pos+max_chars+1]
    text(shown_str, 60, 51)
    cursor_str = " "*cursor_pos + "_"
    text(cursor_str, 60, 51)

def key_pressed():
    global typed_str, str_pos, cursor_pos
    if key == "Backspace" and len(typed_str) > 0:
        backspace()
    elif key == "ArrowLeft":
        arrow_left()
    elif key == "ArrowRight":
        arrow_right()
    elif key == "Delete":
        delete()
    elif len(key) == 1:
        curr_char = str_pos + cursor_pos
        typed_str = typed_str[:curr_char] + key + typed_str[curr_char:]
        if cursor_pos < max_chars:
            cursor_pos += 1
        else:
            str_pos += 1
    elif key == "Enter":
        print(typed_str)
        typed_str = ""
        str_pos = 0
        cursor_pos = 0

def key_repeated():
    global typed_str
    if key == "Backspace" and len(typed_str) > 0:
        backspace()
    elif key == "Delete":
        delete()
    elif key == "ArrowLeft":
        arrow_left()
    elif key == "ArrowRight":
        arrow_right()

def backspace():
    global typed_str, str_pos, cursor_pos
    prev_char = str_pos + cursor_pos - 1
    if prev_char >= 0:
        typed_str = typed_str[:prev_char] + typed_str[prev_char+1:]
    if str_pos > 0:
        str_pos -= 1
    elif cursor_pos > 0:
        cursor_pos -= 1

def delete():
    global cursor_pos
    if cursor_pos + str_pos < len(typed_str):
        cursor_pos += 1
        backspace()
        
def arrow_left():
    global str_pos, cursor_pos
    if str_pos > 0:
        if cursor_pos > 1:
            cursor_pos -= 1
        else:
            str_pos -= 1
    else:
        if cursor_pos > 0:
            cursor_pos -= 1

def arrow_right():
    global str_pos, cursor_pos
    if cursor_pos + str_pos < len(typed_str):
        if cursor_pos < max_chars:
            cursor_pos += 1
        else:
            str_pos += 1
    
```
